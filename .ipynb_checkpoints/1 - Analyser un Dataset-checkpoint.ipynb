### Visualiser la cible

La cible (ou target) est la valeur que nous cherchons à prédire. Dans notre cas, il s'agit du prix d'un billet d'avion représenté par la colonne 'price'. 

```python
plt.figure(figsize=[6, 3])
sns.histplot(data=df, x="price", kde=True)
plt.plot([30000,30000], [40000, 0], ls='--', c='g')
plt.show()
```

On peut observer qu'au dela de 80 000, la courbe s'aplatit fortement. On peut alors considérer ces valeurs comme des outliers.
De plus, on peut observer deux formes en cloche : une vers les 8000 et l'autre vers les 50 000. On voit ainsi que l'on pourrait découper nos données en deux catégories distinctes.

```python
# Retirer les outliers
```

```python
# Créer une variable 'low_price' contenant les données pour les prix inférieurs à 30 000
# Puis créer une variable 'high_price' contenant les données pour les prix supérieurs à 30 000
```

```python
plt.figure(figsize=[17,4])
for n, col in enumerate(numerical_data.iloc[:, 0:3]):
    ax = plt.subplot(1, 3, n+1)
    
    sns.histplot(data=low_price[col], kde=True, label='low price', color='red')
    sns.histplot(data=high_price[col], kde=True, label='high price', color='orange')
    plt.legend()
```

Cette découpe nous permet de mieux observer le comportement du prix en fonction des autres variables. On observe notamment qu'en moyenne les trajets les plus longs coûtent plus chers.
