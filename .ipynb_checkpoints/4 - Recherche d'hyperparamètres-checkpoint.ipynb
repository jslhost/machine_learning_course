# Import

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
```

Une fois que nous avons un modèle (ou une sélection de modèles) qui nous semble prometteur, on peut chercher à optimiser ses hyperparamètres afin de continuer à améliorer nos performances.
Qu'est ce que les hyperparamètres ? 

- Il s'agit de paramètres que notre modèle n'apprend pas au cours de l'entrainement mais que l'on définit en amont par nous même.

Il peut s'agir par exemple du taux d'apprentissage ou du taux de régularisation du modèle.

# Chargement de notre précédent modèle

```python
import joblib
# Charger un modèle enregistré
model = joblib.load('model.pkl')
```

On va réutiliser les fonctions que l'on a écrit précédemment et charger les données concernant les prix des billets d'avion.

```python
def plot_learning_curve(N, train_score, val_score, model_name):
    plt.figure(figsize = (12, 8))
    plt.plot(N, np.absolute(train_score.mean(axis = 1)), label = 'train score')
    plt.plot(N, np.absolute(val_score.mean(axis = 1)), label = 'validation score')
    plt.xlabel("Nombre de données utilisées pour effectuer l'entrainement")
    plt.ylabel("Score (RMSE)")
    plt.legend()
    plt.title(model_name, fontsize=15)
```

```python
def encode_stops(df):
    stops_name = df.value_counts('stops').index
    stops_range = range(len(stops_name))
    stops_dict = dict(zip(stops_name, stops_range))
    stops_dict = {key : str(value) for key, value in stops_dict.items()} # Convert dict value to str
    for key, value in stops_dict.items():
        df['stops'] = df['stops'].apply(lambda x : x.replace(key, value))
    df['stops'] = df['stops'].astype('int64')
    return df
```

```python
def encode_flight(df):
    df['flight'] = df['flight'].apply(lambda x : x[:2])
    return df
```

```python
def encoder(df):
    df = encode_stops(df)
    df = encode_flight(df)
    
    X = df.drop('price', axis=1)
    y = df['price']
    
    print(f"Target mean = {round(y.mean())}")
    
    return X, y
```

```python
# Accélérer l'entrainement
df = pd.read_csv('Clean_Dataset.csv').drop(columns="Unnamed: 0")
df_sample = df.sample(frac=0.03, random_state=4) # sampling to accelerate the training
X_sample, y_sample = encoder(df_sample)
print(df_sample.shape)
```

# Optimisation des hyperparamètres

### GridSearchCV

La première méthode afin de trouver des hyperparamètres optimaux est de faire une recherche par grille avec cross validation.

On va simplement définir dans une grille des hyperparamètres et y associer les valeurs que l'on souhaite tester afin d'obtenir de meilleurs résultats. 

Le modèle va s'entrainer sur l'ensemble des possibilités créés par notre grille, s'évaluer avec une méthode de cross validation puis nous donner la combinaison qui a obtenu les meilleurs résultats. 

C'est une méthode efficace mais qui demande parfois d'avancer à tâtons afin de tester de nouvelles combinaisons. Il s'agit donc d'un outil qu'on utilisera quand on aura suffisament de temps pour essayer d'optimiser notre modèle.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import learning_curve
```

```python
model
```

```python
# On peut définir une grille à l'aide d'un dictionnaire
param_grid = {'sgdregressor__penalty' : ['l1', 'l2'],
              'sgdregressor__alpha' : [0.0001, 0.001, 0.01, 0.1]}
```

```python
# Il suffit de créer un objet GridSearchCV, d'y passer notre modèle et de l'entrainer avec la méthode fit()
grid_search = GridSearchCV(model, param_grid, cv=3, 
                           scoring='neg_root_mean_squared_error',
                           n_jobs=-1)

grid_search.fit(X_sample, y_sample)
```

```python
# On peut ensuite afficher la meilleure combinaison trouvée par notre grille
grid_search.best_params_
```

```python
# Et accéder au meilleur modèle que l'on a entrainé
grid_search.best_estimator_
```

```python
# On peut aussi voir comment se sont débrouillés tous les modèles entrainés durant la recherche
cvres = grid_search.cv_results_

for mean_score, params in zip(cvres['mean_test_score'], cvres['params']):
    print(-mean_score, params)
```

```python
# Enfin, on peut utiliser la learning curve pour évaluer notre modèle
N, train_score, val_score = learning_curve(grid_search.best_estimator_, X_sample, y_sample, 
                                           cv = 5, 
                                           train_sizes = np.linspace(0.1, 1, 10), 
                                           scoring='neg_root_mean_squared_error',
                                           random_state=10)
```

```python
plot_learning_curve(N, train_score, val_score, 'SGD Regressor optimisé par GridSearchCV')

train_score_mean = np.absolute(train_score.mean(axis=1))
val_score_mean = np.absolute(val_score.mean(axis=1))
difference_score = val_score_mean - train_score_mean

my_label = "error difference"
for i, train_val in enumerate(zip(train_score_mean, val_score_mean)):
    plt.plot([N[i], N[i]], [train_val[0], train_val[1]], ls='--', c='r', lw=1)
    plt.scatter(N[i], train_val[0], c='b', zorder=2)
    plt.scatter(N[i], train_val[1], c='g', zorder=2)
    plt.scatter(N[i], train_val[1] - (difference_score[i]/2), c='black', zorder=2, label=my_label)
    plt.annotate(str(round(difference_score[i])),xy=(N[i]+200, train_val[1] - (difference_score[i]/2)-1))
    my_label = "_nolegend_"

plt.legend()
plt.show()
```

### RandomizedSearchCV

Il existe une méthode qui permet de plus facilement gérer le temps que l'on souhaite allouer à la recherche d'hyperparamètres. 

Cette fois, au lieu de tester toutes les combinaisons possibles dans un nombre de valeurs prédéfinies, nous allons donner un intervalle de valeurs dans lequel notre algorithme de recherche va venir piocher aléatoirement durant un nombre d'itérations que l'on va définir.

Cela permet de garder le contrôle sur le nombre de recherche que va effectuer l'algorithme tout en obtenant de bons résultats. Cette méthode peut aussi nous donner une première intuition de la valeur que devrait prendre nos hyperparamètres.
